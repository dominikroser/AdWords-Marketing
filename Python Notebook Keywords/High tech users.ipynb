```python
import advertools as adv
```

```python
kw_df = adv.kw_generate(products=['LG','Sony','Sharp'],
                words=['Ultra HD', '58 inch', '60 inch','65 inch','70 inch','75 inch','85 inch','buy','compare','discount'],
                match_types=['Exact','Phrase','Modified'],
                order_matters=False,
                max_len=2,
                campaign_name='High tech users',
               )
```

```python
kw_df
```

```python
def func(x):
    if 'inch' in x:
        return "size"
    elif 'QLED' in x:
        return "resolution"
    elif 'Ultra HD' in x:
        return "resolution"
    elif 'Tizen' in x:
        return "techie"
    else:
        return 'potential buyer'
    

kw_df['Ad Group'] = kw_df['Keyword'].apply(func)

```

```python
def func1(criterion,keyword):
    if criterion=='Exact':
        return "["+keyword+"]"
    elif criterion=='Broad':
        return keyword
    else:
        return "'"+keyword+"'"
    
    
kw_df['Keyword'] = kw_df.apply(lambda x: func1(x['Criterion Type'], x['Keyword']),axis=1)
```

```python
# Save the final keywords to a CSV file
kw_df.to_excel('high_tech_users_keywords.xlsx',index=False)

# View a summary of our campaign work
summary = kw_df.groupby(['Ad Group', 'Criterion Type'])['Keyword'].count()
summary
```
