```python
import advertools as adv
```

```python
kw_df = adv.kw_generate(products=['TV','electronic TV'],
                words=['discount 40%', 'discount 50%', 'discount 60%','less than 25,000₹', 'big discount','cheap','20,000₹','buy','safe money'],
                match_types=['Exact','Phrase','Modified'],
                order_matters=False,
                max_len=2,
                campaign_name='cost sensitive users',
               )
```

```python
kw_df
```

```python
def func(x):
    if 'inch' in x:
        return "size"
    elif 'discount' in x:
        return "discount"
    else:
        return 'potential buyer'
    

kw_df['Ad Group'] = kw_df['Keyword'].apply(func)

```

```python
def func1(criterion,keyword):
    if criterion=='Exact':
        return "["+keyword+"]"
    elif criterion=='Broad':
        return keyword
    else:
        return "'"+keyword+"'"
    
    
kw_df['Keyword'] = kw_df.apply(lambda x: func1(x['Criterion Type'], x['Keyword']),axis=1)
```

```python
# Save the final keywords to a CSV file
kw_df.to_excel('cost_sensitive_user.xlsx',index=False)

# View a summary of our campaign work
summary = kw_df.groupby(['Ad Group', 'Criterion Type'])['Keyword'].count()
summary
```
